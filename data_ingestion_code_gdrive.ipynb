# Sample reading in data, e.g. all PDFs with tables based on whether they are text-based or image-based

P.S. use your own csv if you downloaded dataset in some other way than the section above

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Omdena_Challenge/new_LK_tea_dataset/v2_new_LK_tea_dataset_updated.csv')
df.fillna(value='', inplace=True)

df.head()
```

Get one sample row from where has_tables is yes and document is text-based and the other is image_based

```python
# Create the filtered and grouped DataFrame
df_filtered = df[df['has_tables'] == 'yes'].groupby('required_ocr')

# Display the grouped DataFrame
df_filtered = df_filtered.apply(lambda x: x)  # Converts the groupby object back to DataFrame for display
df_filtered
```

### checking if files exist where they are supposed to:

```python
import os

def check_files_exist(df):
    """Checks if files exist at the specified paths in the DataFrame."""
    for index, row in df.iterrows():
        pdf_path = row['path']
        text_path = row['text_path']

        # Check PDF file existence
        if pdf_path and os.path.exists(pdf_path):
            print(f"PDF found at: {pdf_path}")
        else:
            print(f"PDF not found at: {pdf_path}")

        # Check text file existence
        if text_path and os.path.exists(text_path):
            print(f"Text file found at: {text_path}")
        else:
            print(f"Text file not found at: {text_path}")

# Assuming df_filtered is your DataFrame containing the paths
check_files_exist(df_filtered)
```

#### all files are present and working on the given paths

## Optional: copy pdfs with tables to another folder with its csv, and create another csv with acts/pdfs without tables and move its files to another folder

```python
import shutil
import os

# Create folders for image_based_PDFs and text_based_PDFs
pdfs_with_tables_folder = os.path.join(local_folder_path, 'PDFs_with_tables')
image_based_pdfs_folder = os.path.join(pdfs_with_tables_folder, 'image_based_PDFs')
text_based_pdfs_folder = os.path.join(pdfs_with_tables_folder, 'text_based_PDFs')
pdfs_without_tables_folder = os.path.join(local_folder_path, 'PDFs_without_tables')


os.makedirs(pdfs_with_tables_folder, exist_ok=True)
os.makedirs(image_based_pdfs_folder, exist_ok=True)
os.makedirs(text_based_pdfs_folder, exist_ok=True)
os.makedirs(pdfs_without_tables_folder, exist_ok=True)

# Copy files based on required_ocr
for index, row in df_filtered.iterrows():
  pdf_path = row['path']
  if pdf_path and os.path.exists(pdf_path):
    if row['required_ocr'] == 'yes':
      shutil.copy(pdf_path, image_based_pdfs_folder)
    else:
      shutil.copy(pdf_path, text_based_pdfs_folder)

# Create CSV for df_filtered
df_filtered.to_csv(os.path.join(pdfs_with_tables_folder, 'PDFs_with_tables.csv'), index=False)

# Create a new DataFrame for rows other than df_filtered
df_remaining = df[~df.filename.isin(df_filtered.filename)]

# Create CSV for df_remaining in PDFs_with_tables folder
df_remaining.to_csv(os.path.join(pdfs_without_tables_folder, 'Acts_and_PDFs_without_tables.csv'), index=False)

# Copy PDFs for df_remaining to PDFs_without_tables folder
for index, row in df_remaining.iterrows():
  pdf_path = row['path']
  if pdf_path and os.path.exists(pdf_path):
    shutil.copy(pdf_path, pdfs_without_tables_folder)
```
